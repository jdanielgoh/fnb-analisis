```python
import geopandas as gpd 
import pandas as pd
import matplotlib.pyplot as plt 
from shapely.geometry import Point, LineString,MultiLineString
from shapely.ops import nearest_points
```

```python
infraestructura_vial = gpd.read_file("datos/infraestructura-vial-ciclista.json")
puntos_accidentes = gpd.read_file("datos/Puntos de accidentes de Ciclistas/accidentado_ciclista.shp")
hechos_transito =pd.read_csv("datos/hechos_transito.csv").drop([107452,107463,107465])
hechos_transito_bicicleta = hechos_transito[hechos_transito.apply(lambda x : "BICICLETA" in [x.tipo_de_vehiculo_1,x.tipo_de_vehiculo_2,x.tipo_de_vehiculo_3,x.tipo_de_vehiculo_4],axis=1)]
hechos_transito_bicicleta.coordenada_y=hechos_transito_bicicleta.coordenada_y.replace(99.183284,-99.183284)
hechos_transito_bicicleta["geometry"]=hechos_transito_bicicleta.apply(lambda x : Point(x.coordenada_y,x.coordenada_x),axis=1)
hechos_transito_bicicleta = gpd.GeoDataFrame(hechos_transito_bicicleta)
```

# Distancias a ciclopistas 


```python
hechos_transito_bicicleta.crs = 'EPSG:4326'
infraestructura_vial.crs = 'EPSG:4326'

hechos_transito_bicicleta = hechos_transito_bicicleta.to_crs('EPSG:32614')
infraestructura_vial = infraestructura_vial.to_crs('EPSG:32614')

def min_distance_to_line(point, line):
    if line.geom_type == 'MultiLineString':
        # Si es una MultiLineString, calcular la distancia para cada parte y tomar la mínima
        min_distances = [point.distance(part) for part in line]
        return min(min_distances)
    elif line.geom_type == 'LineString':
        # Si es una LineString, calcular la distancia directamente
        return point.distance(line)
    else:
        return None  # Manejar otros tipos de geometrías según sea necesario
```

```python
hechos_transito_bicicleta["distancia_minima_infraestructura"]=hechos_transito_bicicleta.apply(lambda row: min_distance_to_line(row.geometry, infraestructura_vial.unary_union), axis=1)
```
